## Checking & Handling Missing Values

```python
#Checking for Categorical missing values 
feature_with_na = [features for features in cyclistic.columns if cyclistic[features].isnull().sum()>1 and cyclistic[features].dtypes == 'O']
feature_with_na

for features in feature_with_na:
    print(features, np.round(cyclistic[features].isnull().mean()* 100 ,2), '% missing values')
```

```python
#Replacing the categorical missing values 

for features in feature_with_na:
    #replace using mode value
    mode_value = cyclistic[features].mode()[0]
    cyclistic[features].fillna(mode_value, inplace = True)

cyclistic[feature_with_na].isnull().sum()
```

```python
#checking for numerical variables missing value
numerical_feat_na = [features for features in cyclistic.columns if cyclistic[features].isnull().sum()>1 and cyclistic[features].dtypes != 'O']
feature_with_na

for features in numerical_feat_na :
    print(features, np.round(cyclistic[features].isnull().mean()* 100 ,2), '% missing values')
```

```python
# dropping the missing values of numerical variables
#making a list of columns to clean
columns_to_clean = ['end_lat', 'end_lng']

#remove rows with missing values from the specified columns
cyclistic.dropna(subset = columns_to_clean, inplace = True)

cyclistic[numerical_feat_na].isnull().sum()
```

```python
# Assuming cyclistic is your DataFrame and 'ride_id' is the column you want to check for duplicates
if 'ride_id' in cyclistic.columns:
    # Check for duplicates in the 'ride_id' column
    duplicates = cyclistic['ride_id'].duplicated()

    # If there are any duplicates, duplicates will contain True for duplicated rows
    if any(duplicates):
        print("There are duplicate values in the 'ride_id' column.")
    else:
        print("No duplicate values found in the 'ride_id' column.")
else:
    print("Column 'ride_id' not found in the DataFrame.")
```

## Adding new columns

```python
# adding new column ride_lenght
#converting to datetime format
cyclistic['started_at'] = pd.to_datetime(cyclistic['started_at'])
cyclistic['ended_at'] = pd.to_datetime(cyclistic['ended_at'])

#calculate ride_lenght
cyclistic["ride_length"] = cyclistic["ended_at"] - cyclistic["started_at"]

#format ride_length to HH:MM:SS
cyclistic["ride_length_formatted"] = cyclistic["ride_length"].dt.total_seconds().apply(lambda x: "{:02}:{:02}:{:02}".format(int(x // 3600), int((x % 3600)// 60), int(x % 60)))

cyclistic.head()
```

```python
cyclistic.drop('ride_length', axis=1, inplace = True)
```

```python
cyclistic.head()
```

```python
cyclistic.info()
```

```python
#adding weekday column
#Create a new column 'day_of_week'
cyclistic['day_of_week'] =(cyclistic['started_at'].dt.dayofweek+1) % 7

#display the resulting df
cyclistic.head()
```

```python
# Create a custom mapping for day of week
day_of_week_mapping = {0: 2, 1: 3, 2: 4, 3: 5, 4: 6, 5: 7, 6: 1}

# Apply the mapping to create a new column 'day_of_week_custom'
cyclistic['day_of_week'] = cyclistic['started_at'].dt.dayofweek.map(day_of_week_mapping)

print(cyclistic[['started_at', 'day_of_week']])
```

```python
cyclistic.shape
```

```python
#create new column month
cyclistic['month'] = cyclistic['started_at'].dt.month
```

```python
cyclistic.head()
```

```python
cyclistic = cyclistic.rename(columns = {'ride_length_formatted' : 'ride_length'})
```

```python
#removing rides less than 3 mins cause ride length is too small
cyclistic = cyclistic[cyclistic['ride_length'] > '00:03:00']
```

```python
cyclistic.shape
```

```python
cyclistic['started_at'].dt.hour
```

```python
#mapping numeric values in the 'day_of_week' column to their corresponding string names
day_of_week_mapping = {1:'Sunday', 2:'Monday', 3:'Tuesday', 4:'Wednesday', 5:'Thursday', 6:'Friday', 7:'Saturday'}

cyclistic['day_of_week'] = cyclistic['day_of_week'].apply(lambda x: day_of_week_mapping[x])
```

```python
#mapping numeric values in the 'month' column to their corresponding string names
month_mapping = {1:'January', 2:'February', 3:'March'}

cyclistic['month'] = cyclistic['month'].apply(lambda x: month_mapping[x])
```

```python
cyclistic.head()
```

```python
cyclistic['ride_duration_mins'] = ((cyclistic['ended_at'] - cyclistic['started_at']).dt.total_seconds() / 60).round(3)

cyclistic.head()
```
